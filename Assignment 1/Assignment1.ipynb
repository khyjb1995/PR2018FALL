```python
import numpy as np
import cv2
from sklearn.naive_bayes import GaussianNB
import os, sys
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from matplotlib.patches import Circle


otherpath = "other/"
duckpath = "duck/"
otherlist = os.listdir(otherpath)
ducklist = os.listdir(duckpath)
piclist = []
labellist = []
    
for fname in otherlist:
    img = cv2.imread(otherpath + fname, 1)
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            piclist.append(img[i,j])
            labellist.append(0)

for fname in ducklist:
    img = cv2.imread(duckpath + fname, 1)
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            piclist.append(img[i,j])
            labellist.append(1)


piclist = np.array(piclist)
labellist = np.array(labellist)

nb = GaussianNB()
nb.fit(piclist, labellist)

result_img = cv2.imread('count02.jpg', 1)
other_img = cv2.imread('count02.jpg', 1)
combine_img = cv2.imread('count02.jpg', 1)

for i in range(result_img.shape[0]):
    temp=[]
    for j in range(result_img.shape[1]) :
        temp.append(result_img[i,j])
        
    temp = np.array(temp)
    predict = nb.predict(temp)
    
    for j in range(result_img.shape[1]) :
        if(predict[j]==1):
            result_img[i,j]=0,0,255
            combine_img[i,j]=0,0,255
        if(predict[j]==0):
            other_img[i,j]=0,0,0
            combine_img[i,j]=0,0,0
        

cv2.imwrite('RedDot_Result.jpg', result_img)
cv2.imwrite('Blackback_Result.jpg', other_img)
cv2.imwrite('Combine_Result.jpg', combine_img)

circle_img = cv2.imread('count02.jpg', 1)
r = 0

l = 12
w = 24
s = 18
appr = 60
sappr = 80
for i in range(other_img.shape[0]):
    for j in range(other_img.shape[1]):
        if (i + s < other_img.shape[0]) and (j + s < other_img.shape[1]):
            count = 0
            if not (other_img[i,j]==(0,0,0)).all():
                for x in range(i,i+s):
                    if count > sappr:
                        break
                    for y in range(j,j+s):
                        if count > sappr:
                            break
                        if (other_img[x,y]==(0,0,0)).all():
                            count = count + 1
                if count < sappr:
                    r = r + 1
                    print("mark" + str(r))
                    for x in range(i,i+s):
                        for y in range(j,j+s):
                            other_img[x,y]=0,0,0
                            #if x < i + 20:
                            circle_img[x,y]=0,0,255
                    cv2.imwrite('output/Circle_Result_'+ str(r) + '.jpg', circle_img)
                    j = j + s
        j = j + 2
                    #cv2.imwrite('output/Circle_Result_Black_'+ str(r) + '.jpg', other_img)
            
for i in range(other_img.shape[0]):
    for j in range(other_img.shape[1]):
        if (i + w < other_img.shape[0]) and (j + l < other_img.shape[1]):
            count = 0
            if not (other_img[i,j]==(0,0,0)).all():
                for x in range(i,i+w):
                    if count > appr:
                        break
                    for y in range(j,j+l):
                        if count > appr:
                            break
                        if (other_img[x,y]==(0,0,0)).all():
                            count = count + 1
                if count < appr:
                    r = r + 1
                    print("mark" + str(r))
                    for x in range(i,i+w):
                        for y in range(j,j+l):
                            other_img[x,y]=0,0,0
                            #if x < i + 20:
                            circle_img[x,y]=0,0,255
                    cv2.imwrite('output/Circle_Result_'+ str(r) + '.jpg', circle_img)
                    j = j + l
        j = j + 2
        
for i in range(other_img.shape[0]):
    for j in range(other_img.shape[1]):
        if (i + l < other_img.shape[0]) and (j + w < other_img.shape[1]):
            count = 0
            if not (other_img[i,j]==(0,0,0)).all():
                for x in range(i,i+l):
                    if count > appr:
                        break
                    for y in range(j,j+w):
                        if count > appr:
                            break
                        if (other_img[x,y]==(0,0,0)).all():
                            count = count + 1
                if count < appr:
                    r = r + 1
                    print("mark" + str(r))
                    for x in range(i,i+l):
                        for y in range(j,j+w):
                            other_img[x,y]=0,0,0
                            #if x < i + 20:
                            circle_img[x,y]=0,0,255
                    cv2.imwrite('output/Circle_Result_'+ str(r) + '.jpg', circle_img)
                    j = j + w
        j = j + 2
            
cv2.imwrite('Circle_Test_Result.jpg', other_img)
cv2.imwrite('Circle_Result.jpg', circle_img)

print("Approximately " + str(r) + " ducks!")
```
